```python
# Импорт библиотек
import numpy as np
import pandas as pd
import keras

# Импорт matplotlib - не нужен, на всякий случай

import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline


# Импорт нужного из keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils

from keras import optimizers
from keras import initializers
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import SGD
```

```python
# Импорт данных
wine = pd.read_csv('winequality-white.csv', sep=';', header=0)
wine.head()
```

```python
#  Разделяем предикторы и отклик
# Отклик - группирующая переменная -  вектор y
y = wine['quality']
# Предикторы - таблица X
X = wine.drop('quality', axis=1)
```

```python
#  Отклик сбалансирован?
wine['quality'].value_counts(normalize=True)
```

```python
#запишем все вина 3й категории в 4ю, а 9й  в 8ю (они похожи по качеству)
for i in range(1,4897):
    if (wine.at[i, 'quality'] == 3):
        wine.at[i, 'quality'] = 4
    if (wine.at[i, 'quality'] == 9):
        wine.at[i, 'quality'] = 8
```

```python
#  Отклик сбалансирован? Теперь да
wine['quality'].value_counts(normalize=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)
```

```python
#  Преобразование pandas dataframe в numpy array
X_train = X_train.values
X_test = X_test.values
y_train = y_train.values
y_test = y_test.values
```

```python
# Больше 2-х классов
# Классы не упорядочены
# Перелываем столбец y в столбцы для каждого класса вина
y_train_bin = np_utils.to_categorical(y_train)

y_test_bin = np_utils.to_categorical(y_test)
```

```python
# Function to create model, required for KerasClassifier
def create_model(optimizer, neurons):
    # default values
    init_mode = initializers.TruncatedNormal(mean=0.0, stddev=0.05, seed=60)
    
    # create model
    model = Sequential()
    model.add(Dense(neurons, 
                    input_dim=11, kernel_initializer=init_mode, 
                    activation='relu')) 
    model.add(Dense(9, activation='softmax'))
    
    # Compile model
    model.compile(loss='binary_crossentropy', 
                  optimizer=optimizer, 
                  metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=create_model, batch_size=10, epochs=100)
```

```python
#momentum_range = [0.0, 0.8] #0.4, 0.6, 0.8, 0.9
#learn_rate_range = [0.001, 0.01] #, 0.1, 0.2, 0.3
neurons_range = [3, 5, 10]
epochs_range = [10, 100]
batch_size_range = [5, 50]
optimizer_range = ['SGD', 'Adam']

#param_grid_test = dict(epochs = epochs_range, batch_size = batch_size_range)
param_grid = dict(epochs = epochs_range, batch_size = batch_size_range, optimizer=optimizer_range, neurons=neurons_range)
grid = GridSearchCV(estimator=model, param_grid = param_grid)

grid_result = grid.fit(X_train, y_train_bin)
```

```python
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
params = grid_result.cv_results_['params']
for mean, param in zip(means, params):
    print("%f with: %r" % (mean, param))
```

```python
model2 = create_model('Adam', 5)
```

```python
# Training a model
model2.fit(X_train, y_train_bin, epochs=100, batch_size=5)
```

```python
# Оценка качества модели
scores = model2.evaluate(X_test, y_test_bin)
print("\nAccuracy: %.2f%%" % (scores[1]*100))
```
