```python
#   Разделяем предикторы и отклик
# Отклик - группирующая переменная -  вектор y
y = wine['Desired1(3)']
# Предикторы - таблица X
X = wine.drop('Desired1(3)', axis=1)
```

```python
обучающей выборки

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=12345,
                                                    # доля объёма тестового множества
                                                    test_size=0.33)
```

```python
#  Преобразование pandas dataframe в numpy array
X_train = X_train.values
X_test = X_test.values
y_train = y_train.values
y_test = y_test.values

```

```python
# Больше 2-х классов
# Классы не упорядочены
y_train_bin = np_utils.to_categorical(y_train)

y_test_bin = np_utils.to_categorical(y_test)
```

```python

i = i +1

# init_2 = initializers.TruncatedNormal(mean=0.0, stddev=0.05, seed=i)
init_2 = initializers.TruncatedNormal(mean=0.0, stddev=0.5, seed=i)
init_3 = initializers.Constant(value = 1e-3)



# Описание архитектуры сети
model = Sequential()
model.add(Dense(9, input_dim=13, activation='relu', kernel_initializer=init_2, bias_initializer=init_3))
model.add(Dense(10, activation='relu', kernel_initializer=init_2, bias_initializer=init_3))
model.add(Dense(3, activation='softmax', kernel_initializer=init_2, bias_initializer=init_3))


# Компиляция модели
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


#  Обучение модели
model.fit(X_train, y_train_bin, epochs=300, batch_size=10)
```

```python
#  Оценка качества модели
scores = model.evaluate(X_test, y_test_bin)
print("\nAccuracy: %.2f%%" % (scores[1]*100))
```

Вместо ручной ре-инициализации пишем бесконечный цикл

Бесконечный цикл очень опасен. Особенно на платных ресурсах, например в облаке.

```python
while True:
    i = i +1

    init_2 = initializers.TruncatedNormal(mean=0.0, stddev=0.5, seed=i)
    init_3 = initializers.Constant(value = 1e-3)

    # Creating a model
    model = Sequential()
    model.add(Dense(9, input_dim=13, activation='relu', kernel_initializer=init_2, bias_initializer=init_3))
    model.add(Dense(10, activation='relu', kernel_initializer=init_2, bias_initializer=init_3))
    model.add(Dense(3, activation='softmax', kernel_initializer=init_2, bias_initializer=init_3))


    # Compiling model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    # Training a model
    model.fit(X_train, y_train_bin, epochs=300, batch_size=10)

    # evaluate the model
    scores = model.evaluate(X_train, y_train)
    print('i=',  i, scores)
    if scores[1]>0.90:
        break
    
    
    
  
```
